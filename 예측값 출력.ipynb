```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import gdown

url='https://docs.google.com/uc?export=download&id=1pmjc3i4Qc6__GqMuwc7DrtrSX0xKsdtC'
gdown.download(url)


```

```python
model=tf.keras.models.load_model('trash_class.h5')
img_path = '/content/paper_011.jpg'
img = image.load_img(img_path, target_size=(384, 384))  # 모델에 맞는 크기로 조정
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)

preds = model.predict(x)
print(preds)

predicted_class = np.argmax(preds[0])

# 예측된 클래스 출력 (클래스 인덱스 or 레이블)
print("Predicted class:", predicted_class)

```
