```python
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np
import gdown

url='https://docs.google.com/uc?export=download&id=1pmjc3i4Qc6__GqMuwc7DrtrSX0xKsdtC'
gdown.download(url)


```

```python
model=tf.keras.models.load_model('trash_class.h5')
img_path = ['/content/cardboard_019.jpg', '/content/metal_032.jpg', '/content/paper_011.jpg','/content/trash_020.jpg','/content/trash_023.jpg']
def load_and_preprocess_images(img_paths):
    imgs = []

    for img_path in img_paths:
        img = image.load_img(img_path, target_size=(384, 384))
        x = image.img_to_array(img)
        x = np.expand_dims(x, axis=0)
        imgs.append(x)

    return np.vstack(imgs)
imgs = load_and_preprocess_images(img_paths)
preds = model.predict(imgs)
for i, pred in enumerate(preds):
    predicted_class = np.argmax(pred)
    print(f"Image {i+1} Predicted class:", predicted_class)


```
